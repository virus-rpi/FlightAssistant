```python
import plotly.graph_objects as go

ay = [x.get("ay") for x in tick_data]
optimal_params = util.find_optimal_savgol_parameters(ay, 10000, 1000)
filtered_ay = util.denoise(ay, optimal_params[0], optimal_params[1])

fig = go.Figure()

fig.add_trace(go.Scatter(x=np.arange(len(ay)), y=ay,
					mode='lines',
					name='ay'))

fig.add_trace(go.Scatter(x=np.arange(len(filtered_ay)), y=filtered_ay,
					mode='lines',
					name='filtered_ay'))

fig.show()
```

```python
ax = [x.get("ax") for x in tick_data]
optimal_params = util.find_optimal_savgol_parameters(ax, 10000, 1000)
filtered_ax = util.denoise(ax, optimal_params[0], optimal_params[1])

fig = go.Figure()

fig.add_trace(go.Scatter(x=np.arange(len(ax)), y=ax,
					mode='lines',
					name='ax'))
fig.add_trace(go.Scatter(x=np.arange(len(filtered_ax)), y=filtered_ax,
					mode='lines',
					name='filtered_ax'))

fig.show()
```
